```python
import geemap.geemap as map
```

```python
import ee
```

```python
ee.Authenticate()
```

```python
ee.Initialize(project = 'ee-nalinapk17')
```

```python
m=map.Map()
m
```

```python
India_Districts = ee.FeatureCollection('projects/ee-nalinapk17/assets/India_Districts')
m.addLayer(India_Districts,{},'India_Districts') #dislpaying the shapefile in the interactive map
m
```

```python
DirectDSR2017 = ee.ImageCollection("MODIS/061/MCD18A1").filterDate("2017-01-01","2017-12-31").select("Direct").mean().clip(India_Districts)
Visualize2017 = {"min":0, "max":1400, 'palette': ['0f17ff', 'b11406', 'f1ff23']}
m.addLayer(DirectDSR2017, Visualize2017, 'DirectDSR_2017')
m
```

```python
import os
```

```python
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
India_DirectDSR2017 = os.path.join(out_dir, 'India_DirectDSR2017.csv')
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
map.zonal_stats(DirectDSR2017, India_Districts, India_DirectDSR2017, statistics_type='MEAN', scale=250)
```
